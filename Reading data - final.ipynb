```python
import numpy as np
import pandas as pd
import os

def my_strip(my_str):
    return my_str.strip()
vstrip = np.vectorize(my_strip)


def vStr2Percent(my_str):
    return float(my_str.replace("%","").strip())


def my_list_joiner(listy,seperator=','):
    return seperator.join(listy)
 
    
def my_spliter(my_array):
    for i in range(len(my_array)):
        my_array[i].split()
vspliter = np.vectorize(my_spliter)

def joiner(dws):
    new_dws = []
    results = ','
    for i in dws:
        if 'more credits' in i :
            pass
        elif 'See full cast & crew' in i:
            pass
        else:
            new_dws.append(i)
    return ','.join(new_dws)

def my_spliter(my_array):
    final_array = []
    for i in range(len(my_array)):
        if (my_array[i] != '-'):
            final_array.append(my_array[i].split())
        elif (my_array[i] == '-'):
            final_array.append(np.NaN)
    
    return final_array        

def table_spread(i, data_series):
    try:
        rows = pd.DataFrame(eval(data_series.values[i])).drop(columns=['Unnamed: 0']).to_numpy()
        for row_i in range(len(rows)):
            rows[row_i] = my_spliter(rows[row_i])  
        return rows
        
        
    
    except:
        return np.NaN
    
```

```python
def related_movie_score(RSDs):
    RSDs = eval(RSDs)
    number_of_votes = []
    score = []
    movie_code = []
    
    for rsd in RSDs:
        try:
            number_of_votes.append(int(rsd[0][0][0]))
        except:
            number_of_votes.append(np.NaN)
        try:
            score.append(float(rsd[0][0][1]))
        except:
            score.append(np.NaN)

        # I did'nt use it
        try:
            movie_code.append(rsd[1])
        except:
            movie_code.append(np.NaN)
            
    return np.array(list(zip(number_of_votes, score)))
```

# Voting rating disro  - expanding data into column

```python
VD = []
for i in range(len(data['vote_distro'])):
    try:
        vote_distro_percent = pd.DataFrame(eval(data['vote_distro'].values[i]))['Unnamed: 1'].apply(vStr2Percent).to_numpy()
        vote_distro_num = pd.DataFrame(eval(data['vote_distro'].values[i]))['Votes'].to_numpy()
        VD.append(list(zip(vote_distro_percent, vote_distro_num)))
    except:
        VD.append(np.NaN)
```

```python
rating_distro = {}
l = []
for rating in range(10):
    for j in range(len(VD)):
        try:
            l.append(VD[j][rating])
        except:
            l.append(np.NaN)
    rating_distro['Rating %d (Percent , Vote Num)'%(10 - rating)] = l
    l = []
```
