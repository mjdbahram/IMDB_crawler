# Opening Weekend USA

```python
def opening_week_unit(my_str):
    try:
        budget = my_str.split()
        unit = budget[0][0]
        return unit
    except:
        return np.nan
    
def opening_week_val(my_str):
    try:
        budget = my_str.split()
        value = int((budget[0][1:]).replace(',',''))
        return value
    except:
        return np.nan
    

def opening_week_date(my_str):
    try:
        date = re.findall(r'(\d+ \w+ \d+)',my_str)
        return datetime.strptime(date[0], '%d %B %Y')
    except:
        return np.nan
    
    
def opening_week_rel(my_str):
    try:
        wr = re.findall(r'\s[^0-9]\w*\s[^0-9]\w+',my_str)
        return wr[0]
    except:
        return np.nan
    
    
```

```python
def pop_u_d(my_str):
    try:
        my_str = my_str.split()
        if (my_str[0]=='Up'):
            return int(my_str[1])
        elif (my_str[0]=='Down'):
            return -1*int(my_str[1])
        else:
            return 'mj'
        
    except:
        return np.nan
        
```

# 'Related movie'

```python
def rel_mvs_name(rel_mv):
    try:
        result = []
        rel_mv = eval(rel_mv)
        for mov in rel_mv:
            result.append(mov[0])
        return ','.join(result)
    except:
        return np.NaN
        
    
```

```python
def rel_mvs_id(rel_mv):
    try:
        result = []
        rel_mv = eval(rel_mv)
        for mov in rel_mv:
            result.append(mov[1].split('/')[-2])
        return ','.join(result)
    except:
        return np.NaN
```
